```python
X = df[['gdp_per_capita','health','freedom','cpi_score']]
```

```python
y = df['happiness_score']
```

```python
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
r2 = model.score(X,y)
print(r2)
```

```python
print('intercept is ',model.intercept_)
print('coefficients is ',list(zip(X, model.coef_)), sep='\n')
```
